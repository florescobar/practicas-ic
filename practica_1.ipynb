# Prácticas 1. Redes neuronales. Perceptrón

```python
from keras.datasets import mnist

(train_X, train_y), (test_X, test_y) = mnist.load_data()
```

```python
print('X_train: ' + str(train_X.shape))
print('Y_train: ' + str(train_y.shape))
print('X_test:  '  + str(test_X.shape))
print('Y_test:  '  + str(test_y.shape))
```

```python
from matplotlib import pyplot as plt
from itertools import product
import numpy as np

nrows = 5
ncols = 5

fig, ax = plt.subplots(nrows, ncols, figsize=(20,20))

for i, j in product(range(nrows), range(ncols)):
    ax[i, j].imshow(train_X[np.random.randint(0, train_X.shape[0]) ], cmap=plt.get_cmap('gray'))


plt.show()
```

```python
from sklearn.linear_model import Perceptron

x_train = train_X.astype("float64")/255.0
x_train = x_train.reshape((60000, (28*28)))

x_test = test_X.astype("float64")/255.0
x_test = x_test.reshape((10000, (28*28)))


p = Perceptron(random_state=32, max_iter=50, tol=0.0001)

p.fit(X=x_train, y=train_y)

error_rate = (1 - p.score(X=x_test, y=test_y)) * 100

print(error_rate)
```
